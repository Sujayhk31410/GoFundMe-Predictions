```python
import numpy as np
import pandas as pd
import requests
import time

# Initialize API key, secret, and detection URL
api_key = "O0Jv9C_KEfhKHHlETYZyNxCFOXSopkz4"
api_secret = "FoeSGIifkhlYetmVmjLIQWOoXrOc3XHE"
detect_url = 'https://api-us.faceplusplus.com/facepp/v3/detect'

# Load dataset
df = pd.read_csv('processed_data.csv')

# Filter for medical category
medical_df = df[df['category'] == 'medical-fundraiser']
medical_sample = medical_df.sample(frac=0.8,random_state=1)  # Set random_state for reproducibility

# Create lists to store results
image_urls = []
genders = []
ages = []
smiles = []
emotions = []
beauty_scores = []

# Loop through each image in the sampled medical category
for index, row in medical_sample.iterrows():
    image_url = row['image_url']
    
    # Prepare the detection parameters
    detect_params = {
        'api_key': api_key,
        'api_secret': api_secret,
        'image_url': image_url,
        'return_attributes': 'gender,age,smiling,emotion,beauty'  # Customize attributes
    }

    try:
        # Make the request to detect faces
        detect_response = requests.post(detect_url, data=detect_params)
        detect_data = detect_response.json()
        
        # Check if faces are detected
        if 'faces' in detect_data and len(detect_data['faces']) > 0:
            # Extract the face attributes
            face_data = detect_data['faces'][0]['attributes']
            
            # Store the relevant data
            image_urls.append(image_url)
            genders.append(face_data['gender']['value'])
            ages.append(face_data['age']['value'])
            smiles.append(face_data['smile']['value'])
            emotions.append(face_data['emotion'])
            beauty_scores.append({
                'female_score': face_data['beauty']['female_score'],
                'male_score': face_data['beauty']['male_score']
            })
        else:
            print(f"No faces detected in image at {image_url}.")
    
    except Exception as e:
        print(f"Error processing image at {image_url}: {e}")
    
    # Pause to avoid hitting API rate limits
    time.sleep(1)

# Create a DataFrame to store the results
results_df = pd.DataFrame({
    'image_url': image_urls,
    'gender': genders,
    'age': ages,
    'smile': smiles,
    'emotion': emotions,
    'beauty': beauty_scores
})

# Save the DataFrame to a CSV file
results_df.to_csv('face_detection_medical_results.csv', index=False)

# Print a preview of the results
print(results_df.head())
```
